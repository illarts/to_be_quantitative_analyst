Á¢∫ÁéáË´ñ„ÉªÁ¢∫ÁéáÈÅéÁ®ã„ÉªÈáëËûçÂ∑•Â≠¶„ÅÆÊ¶ÇÂøµ„Çí „Éó„É≠„Ç∞„É©„Éü„É≥„Ç∞ÔºàPythonÔºâ „Å´ËêΩ„Å®„ÅóËæº„ÇÄÊñπÊ≥ï„Çí„ÄÅ‰ª•‰∏ã„ÅÆÊµÅ„Çå„ÅßËß£Ë™¨„Åó„Åæ„Åô„ÄÇ

üîπ 1. „Éó„É≠„Ç∞„É©„Éü„É≥„Ç∞„ÅÆÁõÆÁöÑ„ÇíÊòéÁ¢∫„Å´„Åô„Çã

„Åæ„Åö„ÄÅÊï∞Â≠¶„ÅÆÊ¶ÇÂøµ„Çí„Éó„É≠„Ç∞„É©„É†„ÅßÊâ±„ÅÜÁõÆÁöÑ„ÇíËÄÉ„Åà„Åæ„Åô„ÄÇ
ÈáëËûçÂ∑•Â≠¶„ÅÆË¶ñÁÇπ„Åã„Çâ„ÅØ„ÄÅÊ¨°„ÅÆ„Çà„ÅÜ„Å™ÁõÆÁöÑ„ÅåËÄÉ„Åà„Çâ„Çå„Åæ„Åô„ÄÇ




```python
import numpy as np

# Ê®ôÊ∫ñÊ≠£Ë¶èÂàÜÂ∏É„Å´Âæì„ÅÜ‰π±Êï∞„Çí10ÂÄãÁîüÊàê
random_numbers = np.random.normal(0, 1, 100)
print(random_numbers)
```

# „É¢„É≥„ÉÜ„Ç´„É´„É≠„Ç∑„Éü„É•„É¨„Éº„Ç∑„Éß„É≥

```python
def monte_carlo_simulation(n):
    samples = np.random.normal(0, 1, n)  # Ê≠£Ë¶èÂàÜÂ∏É„Åã„ÇânÂÄã„ÅÆ‰π±Êï∞„ÇíÂèñÂæó
    return np.mean(samples)  # ÊúüÂæÖÂÄ§„ÇíË®àÁÆó

print(monte_carlo_simulation(10000))
```

```python
import matplotlib.pyplot as plt

# „Éñ„É©„Ç¶„É≥ÈÅãÂãï„ÅÆ„Ç∑„Éü„É•„É¨„Éº„Ç∑„Éß„É≥
def brownian_motion(T=1, N=1000):
    dt = T/N  # ÊôÇÈñìÂàª„Åø
    dW = np.random.normal(0, np.sqrt(dt), N)  # Áã¨Á´ã„Å™Ê®ôÊ∫ñÊ≠£Ë¶è‰π±Êï∞
    W = np.cumsum(dW)  # Á¥ØÁ©çÂíå„ÇíÂèñ„Å£„Å¶„Éñ„É©„Ç¶„É≥ÈÅãÂãï„Çí‰ΩúÊàê
    return W

N = 1000
T = 1
W = brownian_motion(T, N)

plt.plot(np.linspace(0, T, N), W)
plt.xlabel("Time")
plt.ylabel("Wiener Process")
plt.title("Simulation of Brownian Motion")
plt.show()
```

# Á¢∫ÁéáÂæÆÂàÜÊñπÁ®ãÂºèÔºàSDEÔºâ„ÅÆÊï∞ÂÄ§Ëß£Ê≥ï

üìå ‚ë£ Euler-MaruyamaÊ≥ï„ÅßSDE„ÇíËß£„Åè

```python
def euler_maruyama(S0, mu, sigma, T, N):
    dt = T/N
    S = np.zeros(N)
    S[0] = S0
    for i in range(1, N):
        dW = np.random.normal(0, np.sqrt(dt))  # „Éñ„É©„Ç¶„É≥ÈÅãÂãï„ÅÆÂ§âÂåñÈáè
        S[i] = S[i-1] + mu * S[i-1] * dt + sigma * S[i-1] * dW
    return S

# ÂàùÊúüÂÄ§
S0 = 100
mu = 0.05
sigma = 0.2
T = 1
N = 1000

# „Ç∑„Éü„É•„É¨„Éº„Ç∑„Éß„É≥
S = euler_maruyama(S0, mu, sigma, T, N)

plt.plot(np.linspace(0, T, N), S)
plt.xlabel("Time")
plt.ylabel("Stock Price")
plt.title("Euler-Maruyama Simulation of Stock Price")
plt.show()
```

# „Éñ„É©„ÉÉ„ÇØ„Éª„Ç∑„Éß„Éº„É´„Ç∫ÊñπÁ®ãÂºè„ÅÆÂÆüË£Ö

üìå ‚ë§ „Éñ„É©„ÉÉ„ÇØ„Éª„Ç∑„Éß„Éº„É´„Ç∫„É¢„Éá„É´

„Ç™„Éó„Ç∑„Éß„É≥‰æ°Ê†º„ÅÆËß£ÊûêËß£„ÇíÊ±Ç„ÇÅ„Çã„ÄÇ

```python
from scipy.stats import norm

def black_scholes(S, K, T, r, sigma, option_type="call"):
    d1 = (np.log(S/K) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)

    if option_type == "call":
        price = S * norm.cdf(d1) - K * np.exp(-r*T) * norm.cdf(d2)
    else:  # put option
        price = K * np.exp(-r*T) * norm.cdf(-d2) - S * norm.cdf(-d1)
    
    return price

# ÂàùÊúü„Éë„É©„É°„Éº„Çø
S = 100   # ÁèæÂú®„ÅÆÊ†™‰æ°
K = 100   # Ê®©Âà©Ë°å‰Ωø‰æ°Ê†º
T = 1     # Ê∫ÄÊúüÔºàÂπ¥Âçò‰ΩçÔºâ
r = 0.05  # ÁÑ°„É™„Çπ„ÇØÈáëÂà©
sigma = 0.2  # „Éú„É©„ÉÜ„Ç£„É™„ÉÜ„Ç£

# „Ç™„Éó„Ç∑„Éß„É≥‰æ°Ê†ºË®àÁÆó
call_price = black_scholes(S, K, T, r, sigma, "call")
put_price = black_scholes(S, K, T, r, sigma, "put")

print(f"Call Option Price: {call_price:.2f}")
print(f"Put Option Price: {put_price:.2f}")
```
