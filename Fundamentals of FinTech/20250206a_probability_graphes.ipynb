

各分布の確率密度関数（PDF）または確率質量関数（PMF）をPythonで可視化します。
以下の分布をMatplotlibとSciPyを使ってプロットします。

- 二項分布（Binomial）
- 一様分布（Uniform）
- ポアソン分布（Poisson）
- 正規分布（Normal）
- コーシー分布（Cauchy）
- 指数分布（Exponential）

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# 図の設定
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

# 1. 二項分布 (Binomial Distribution)
n, p = 10, 0.5  # 試行回数10, 成功確率0.5
x = np.arange(0, n+1)
y = stats.binom.pmf(x, n, p)
axes[0].bar(x, y, alpha=0.6, color='b', label=f'Binomial(n={n}, p={p})')
axes[0].set_title("Binomial Distribution")
axes[0].set_xlabel("x")
axes[0].set_ylabel("PMF")
axes[0].legend()

# 2. 一様分布 (Uniform Distribution)
a, b = 0, 10  # 範囲 [0, 10]
x = np.linspace(a, b, 100)
y = stats.uniform.pdf(x, a, b-a)
axes[1].plot(x, y, 'g-', label=f'Uniform({a}, {b})')
axes[1].set_title("Uniform Distribution")
axes[1].set_xlabel("x")
axes[1].set_ylabel("PDF")
axes[1].legend()

# 3. ポアソン分布 (Poisson Distribution)
lambda_ = 4  # 平均 λ=4
x = np.arange(0, 15)
y = stats.poisson.pmf(x, lambda_)
axes[2].bar(x, y, alpha=0.6, color='r', label=f'Poisson(λ={lambda_})')
axes[2].set_title("Poisson Distribution")
axes[2].set_xlabel("x")
axes[2].set_ylabel("PMF")
axes[2].legend()

# 4. 正規分布 (Normal Distribution)
mu, sigma = 0, 1  # 平均0, 標準偏差1
x = np.linspace(-4, 4, 100)
y = stats.norm.pdf(x, mu, sigma)
axes[3].plot(x, y, 'b-', label=f'Normal(μ={mu}, σ={sigma})')
axes[3].set_title("Normal Distribution")
axes[3].set_xlabel("x")
axes[3].set_ylabel("PDF")
axes[3].legend()

# 5. コーシー分布 (Cauchy Distribution)
x0, gamma = 0, 1  # 中心0, スケール1
x = np.linspace(-5, 5, 100)
y = stats.cauchy.pdf(x, x0, gamma)
axes[4].plot(x, y, 'm-', label=f'Cauchy(x0={x0}, γ={gamma})')
axes[4].set_title("Cauchy Distribution")
axes[4].set_xlabel("x")
axes[4].set_ylabel("PDF")
axes[4].legend()

# 6. 指数分布 (Exponential Distribution)
lambda_ = 1  # λ=1
x = np.linspace(0, 5, 100)
y = stats.expon.pdf(x, scale=1/lambda_)
axes[5].plot(x, y, 'c-', label=f'Exponential(λ={lambda_})')
axes[5].set_title("Exponential Distribution")
axes[5].set_xlabel("x")
axes[5].set_ylabel("PDF")
axes[5].legend()

# レイアウト調整
plt.tight_layout()
plt.show()
```
